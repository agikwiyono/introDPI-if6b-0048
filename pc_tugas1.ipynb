```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

img_bgr = cv2.imread('downloads.jpg') # Pastikan file gambar.jpg ada di folder yang sama
img_bgr = cv2.resize(img_bgr, (300, 450))

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

print("=== INFO DIMENSI GAMBAR ===")
print(f"Dimensi Grayscale : {img_gray.shape}")
print(f"Dimensi RGB       : {img_rgb.shape}\n")

y1, y2, x1, x2 = 200, 208, 180, 188
matrix_gray = img_gray[y1:y2, x1:x2]
matrix_rgb = img_rgb[y1:y2, x1:x2]

print("=== MATRIKS f(x,y) GRAYSCALE 8x8 ===")
print(matrix_gray, "\n")

print("=== MATRIKS f(x,y) RGB 8x8 ===")
print(matrix_rgb, "\n")

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(img_rgb)
ax[0].set_title("Citra Asli (RGB)")
ax[0].axis('off')
ax[0].add_patch(patches.Rectangle((x1, y1), 8, 8, linewidth=2, edgecolor='red', facecolor='none'))

ax[1].imshow(matrix_gray, cmap='gray', vmin=0, vmax=255)
ax[1].set_title("Zoom 8x8 (Grayscale)")
ax[1].axis('off')
for i in range(8):
    for j in range(8):
        val = matrix_gray[i, j]
        color = 'white' if val < 128 else 'black'
        ax[1].text(j, i, str(val), ha='center', va='center', color=color, fontsize=8)

ax[2].imshow(matrix_rgb)
ax[2].set_title("Zoom 8x8 (RGB)")
ax[2].axis('off')
for i in range(8):
    for j in range(8):
        r, g, b = matrix_rgb[i, j]
        lum = (0.299 * r) + (0.587 * g) + (0.114 * b)
        color = 'white' if lum < 128 else 'black'
        ax[2].text(j, i, f"{r}\n{g}\n{b}", ha='center', va='center', color=color, fontsize=7)

plt.tight_layout()
plt.show()
```
